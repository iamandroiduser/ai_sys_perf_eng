## GPU architecture
- CPUs optimize low-latency, single-thread performance
- GPUs optimize throughput, so can run 1000s of threads in parallel
- 1 GPU has many SMs (optimized for throughput)
    - 1 SM can track upto 64 warps (= 2048 threads) concurrently
    - Blackwell has 64K 32-bit registers per SM (= 256 KB), and combined 256 KB L1 cache/shared memory per SM
        - 228 KB of that SRAM can be use-manged shared memory (227 KB usable, 1 KB reserved by CUDA)
        - 1 thread block can request upto 227 KB dynamic shared memory

# Takeaways
## Measure goodput
- beyond FLOPS or utilization
- ratio: 
    - time spent by GPU doing useful stuff (fwd/back prop)
    - time spent by PGU waiting on data / other overhead
- How to measure goodput?
    - Nvidia Nsight Systems/Compute
    - PyTorch profiler
- We need to ^ this ratio

## Incremental Optimizations
- At scale, small % can mean big savings

## Profile-driven mindset
- Use data and pofiling to guide optimization
- Profile to ID bottlenecks: compute utilization/ memory bendwidth/ memory latency/ cache misses/ comm/network delays
- Do targeted optimization for eahc bottlenck IDed

## Holistic View
- Many layers: HW (CPU. GPU. Memory. NW). SW (alog, libs)
    - any of them can be weak
- sometimes tweak HW, sometime SW
